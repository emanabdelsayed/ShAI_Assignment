### **2- Descriptive Statistics**

```python
df.describe()
```

```python
# Calculate and display mean and median for 'TotalPay' column
mean_salary = df['TotalPay'].mean()
median_salary = df['TotalPay'].median()

print(f"Mean Salary: {mean_salary}")
print(f"Median Salary: {median_salary}")

# Calculate and display mode for 'TotalPay' column
mode_total_pay = df['TotalPay'].mode()
print(f"\nMode of TotalPay: {mode_total_pay}")

# Calculate and display minimum and maximum salaries
min_salary = df['TotalPay'].min()
max_salary = df['TotalPay'].max()
print(f"\nMinimum Salary: {min_salary}")
print(f"Maximum Salary: {max_salary}")

# Calculate and display the range of salaries
salary_range = max_salary - min_salary
print(f"\nRange of Salaries: {salary_range}")

# Calculate and display the standard deviation of 'TotalPay' column
std_dev_salary = df['TotalPay'].std()
print(f"\nStandard Deviation of TotalPay: {std_dev_salary}")
```

### 3- Data **Cleaning**

```python
# first drop two categorical columns ['Notes', 'Status'] that have null values
# Because they will not affect the data analysis
df = df.drop(['Notes', 'Status'], axis=1)
```

```python
# second drop row that contains null values they just 4 rows and will not affect to analysis
df.dropna(subset=['OvertimePay', 'OtherPay'], inplace=True)
```

```python
#then replace null here with mean value
df['Benefits'].fillna(df['Benefits'].mean(), inplace=True)
df['BasePay'].fillna(df['BasePay'].mean(), inplace=True)
```

```python
df.isnull().sum()
```

### **4- Basic Data Visualizatio**n

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Visualize the distribution of salaries using a histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['TotalPay'], bins=30, kde=True)
plt.title('Distribution of Salaries')
plt.xlabel('Total Pay')
plt.ylabel('Frequency')
plt.show()
```

```python
# Create a pie chart to represent the proportion of employees in different departments
department_counts = df['JobTitle'].value_counts().head(10)

plt.figure(figsize=(12, 8))
plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Employees in Different Departments')
plt.show()
```

### **5- Grouped Analysis**

```python
# Group by 'Year' and calculate the mean for each numeric column
grouped_by_year = df.groupby('Year').mean()

# Display the result
print(grouped_by_year)
```

```python
# Group by 'JobTitle' and calculate average salary for each group
average_salary_by_job = df.groupby('JobTitle')['TotalPay'].mean()

# Group by 'Year' and 'JobTitle' and calculate average salary for each group
average_salary_by_year_and_job = df.groupby(['Year', 'JobTitle'])['TotalPay'].mean()

# Display the results
print("Average Salary by Job Title:")
print(average_salary_by_job)

print("\nAverage Salary by Year and Job Title:")
print(average_salary_by_year_and_job)
```

### **6- Simple Correlation Analysis**

```python
# Identify correlation between 'TotalPay' and another numerical column
correlation_column = 'BasePay'

correlation = df['TotalPay'].corr(df[correlation_column])

# Plot scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df[correlation_column], df['TotalPay'], alpha=0.5)
plt.title(f'Scatter Plot of TotalPay vs {correlation_column} (Correlation: {correlation:.2f})')
plt.xlabel(correlation_column)
plt.ylabel('TotalPay')
plt.show()
```
